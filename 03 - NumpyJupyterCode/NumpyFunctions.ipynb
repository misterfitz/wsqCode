## Universal Functions
- Function that performs elementwise operations on data in arrays

```python
import numpy as np

# many ufuncs operate on a single array
arr = np.array([1,4,9,16])
np.sqrt(arr)
```

```python
np.log(arr)
```

```python
# some ufuncs operate on multiple arrays 
arr1=np.array([1,-1,3,5])
arr2=np.array([2,1,4,1])
np.maximum(arr1,arr2)
```

### Math/Stat Methods

```python
list1 = [1,1.1,-5.45,3.2]
arr1 = np.array(list1)

list2 = [[1,1.1,-5.45,3.2],[-0.5,0.33,10.12,-5]]
arr2 = np.array(list2)
```

```python
# arrays have many built in functions that aggregate the data 
arr1.mean()
```

```python
print (arr2.mean())
print (arr2.mean(0))
print (arr2.mean(1))
```

```python
# you can call these functions using top-level numpy functions too
np.mean(arr2)
```

```python
# some functions don't aggregate but instead produce an array
arr2.cumsum(0)
```
