# Linear Algebra

```python
import numpy as np
x = np.array([[1., 2., 3.], [4., 5., 6.]])
x
```

```python
y = np.array([[6., 23.], [-1, 7], [8, 9]])
y
```

```python
# multiply matrices
x.dot(y)
```

```python
# can use top-level numpy function
np.dot(x,y)
```

```python
# multiply shorthand
x @ y
```

```python
# transpose
x.T
```

```python
# inverse
np.linalg.inv(x @ y)
```
